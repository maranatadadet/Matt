```python
import numpy as np
from random import random 
import matplotlib.pyplot as plt
print(                      'Problem 1: Monte Carlo Distribution'     )
def function(sig,Zo): 
    r = np.sqrt( -2*(np.log(1- random())) )
    theta = 2*np.pi*random()
    x = r*np.cos(theta)
    return x*sig + Zo

n=[]
for d in range(1000):
    n.append(function(2,-1))
    #print(d, function(2,-1))
    #print(d,n)
    
print(n)
plt.hist(n)

    

```
