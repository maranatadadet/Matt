```python
print('Three doors case')
print()
import numpy as np
import random
win_withswitch = 0
lose_withswitch = 0
win_withoutswitch = 0
lose_withoutswitch = 0
switchgames = 0
noswitchgames = 0
#from random import random.choice
for i in range(0,1000):
    doors = ['door1','door2','door3']
    prize = random.choice(doors)
    pick = random.choice(doors)
    #print('pick and prize :' ,pick, prize)
    
    if prize == pick:
        switch = random.random()
        if switch < 0.5:
            win_withoutswitch +=1
            noswitchgames +=1
            #print("win - noswitch")
        else:
            lose_withswitch += 1
            switchgames += 1
            #print("lose - switch")

    else:
        switch = random.random()
        if switch > 0.5:
            win_withswitch +=1
            switchgames +=1
            #print("win - switch")
        else:
            lose_withoutswitch += 1
            noswitchgames += 1
            #print("lose - noswitch")      
        
        if prize == 'door1' and pick == 'door2':
            doors.remove('door3')
                   
        if prize == 'door2' and pick == 'door1':
            doors.remove('door3')
                
        if prize == 'door2' and pick == 'door3':
            doors.remove('door1')
            
        if prize == 'door3' and pick == 'door2':
            doors.remove('door1')
            
        if prize == 'door1' and pick == 'door3':
            doors.remove('door2')
            
        if prize == 'door3' and pick == 'door1':
            doors.remove('door2')
            
        if prize == 'door1' and pick == 'door1':
            doors.remove('door2')
            
        if prize == 'door2' and pick == 'door2':
            doors.remove('door1')
            
        if prize == 'door3' and pick == 'door3':
            doors.remove('door2')            
            
print('switchgames',switchgames)
print('noswitchgames',noswitchgames)
print('win_withswitch',win_withswitch)            
print('lose_withswitch',lose_withswitch)             
print('win_withoutswitch',win_withoutswitch)
print('lose_withoutswitch',lose_withoutswitch)
print()
print('Switching probability:', win_withswitch/switchgames)
print('No Switching Probablity:', win_withoutswitch/noswitchgames)

      
            
            
            

        
    

        
    
   
    
```
